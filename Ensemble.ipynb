# Ensembe modeling 
    Combing the results of multiple models to get final result 
# Models 
    homogenious : where all the models are of same kind ( for eg use only logistic or decision Tree) 
    Hetrogenious : where all or few models are different  (for eg few logistic and few Decision Tree)
# Combing the results  ( Check -How to combine models)
    Voting : count the freq of each class and produce the final result of major class 
    Average : Taking average of each results of all models
